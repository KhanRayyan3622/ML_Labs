```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme()

heart_data = pd.read_csv('heart.csv')

sns.relplot(
    data=heart_data,
    x="age", y="serum_creatinine", hue="sex",
    size="platelets", sizes=(10, 200),
    style="smoking", col="diabetes"
)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

heart_data = pd.read_csv('heart.csv')
sns.histplot(data=heart_data, x='age', kde=True)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age in Heart Failure Dataset')
plt.show()
```

# Task 4

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'
dataframe = pd.read_csv(url, header=None, na_values='?')

data = dataframe.values
ix = [i for i in range(data.shape[1]) if i != 23]
X, y = data[:, ix], data[:, 23]

print('Missing values before imputation:', np.sum(np.isnan(X).flatten()))

imputer = SimpleImputer(strategy='mean')
Xtrans = imputer.fit_transform(X)

print('Missing values after imputation:', np.sum(np.isnan(Xtrans).flatten()))
```

# Task 5

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

file_path = 'heart.csv'
dataframe = pd.read_csv(file_path)
print('Missing values before imputation:', dataframe.isnull().sum().sum())
imputer = SimpleImputer(strategy='mean')
Xtrans = imputer.fit_transform(dataframe)
print('Missing values after imputation:', np.sum(np.isnan(Xtrans).flatten()))
```
